# 9. 데이터 선택(조건)
조건에 해당하는 데이터 선택. 조건으로 행 필터링 하기. 열로 필터 걸고 행을 반환

```python
df['키'] >= 185 #학생들의 키가 185 이상인지 여부를 알려줌
```

```python
filt = (df['키'] >= 185)
df[filt]
```

```python
df[~filt] #제외한 나머지
```

```python
df[df['키'] >= 185] #filt라는 객체 사용 대신에 
```

```python
df.loc[df['키'] >= 185,'수학'] #키가 185 이상인 학생들의 수학 데이터
```

```python
df.loc[df['키'] >= 185,['이름','수학','과학']] #키가 185 이상인 학생들의 이름, 수학, 과학 데이터
```

## 9.1 다양한 조건

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '용산고')] # 키가 185 이상인 북산고 학생 데이터
```

# 13.1 데이터에 함수 적용하기(apply)

```python
df = pd.read_excel('score.xlsx', index_col ='지원번호')
df
```

```python
# 키 뒤에 cm을 붙이는 역할
def add_cm(height):
    return str(height) + 'cm'

df['키'] = df['키'].apply(add_cm) #키 데이터에 대해서 add_cm 함수를 호출한 결과 데이터를 반영
df
```

```python
def capitalize(lang):
    if pd.notnull(lang): #NaN이 아닌지
        return lang.capitalize() #첫 글자는 대문자로, 나머지는 소문자로
    return lang

df['SW특기'] = df['SW특기'].apply(capitalize)
df
```

```python
df['SW특기'] = df['SW특기'].str.capitalize() #실무에서는 그냥 이거 쓰자
df
```
